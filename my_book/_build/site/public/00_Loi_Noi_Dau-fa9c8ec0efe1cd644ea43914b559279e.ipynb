# LỜI NÓI ĐẦU

Trong kỷ nguyên số hóa hiện đại, sự chuyển dịch vĩ đại của phân tích dữ liệu trong khối ngành kinh tế, tài chính và quản trị kinh doanh đã tạo ra một cuộc cách mạng sâu sắc về mặt phương pháp luận. Kỷ nguyên của những bộ dữ liệu mỏng manh và tư duy phân tích đơn biến (univariate analysis) đã hoàn toàn trở nên lỗi thời, nhường chỗ cho một hệ sinh thái dữ liệu đa chiều, phức tạp và khổng lồ. Ngày nay, mọi quyết định kinh doanh mang tầm vóc chiến lược, từ việc đánh giá rủi ro tín dụng của một hệ thống ngân hàng lớn, dự báo doanh thu của một tập đoàn bán lẻ đa quốc gia, cho đến việc tối ưu hóa chuỗi cung ứng toàn cầu, đều đòi hỏi những bộ óc có khả năng xử lý, mô hình hóa và bóc tách hàng trăm, thậm chí hàng ngàn biến số tương tác chéo lẫn nhau. Cuốn sách này ra đời với một sứ mệnh đặc biệt và một triết lý cốt lõi vô cùng rõ ràng: Trở thành một cây cầu nối thực tiễn, vững chắc và toàn diện giữa hai nhánh tư tưởng vĩ đại nhất của khoa học dữ liệu hiện đại, đó là Suy diễn nhân quả (Statistical Inference) và Học máy hiện đại (Data-driven/Predictive Modeling). Trong khi Suy diễn nhân quả truyền thống của kinh tế lượng đặt trọng tâm vào việc đi tìm câu trả lời minh bạch cho câu hỏi "Tại sao" (Ví dụ: Tại sao khách hàng lại rời bỏ dịch vụ?), đòi hỏi sự diễn giải khắt khe về các cơ chế tác động và kiểm soát chặt chẽ các biến nhiễu thông qua lý thuyết kinh tế, thì Học máy hiện đại lại khao khát tối ưu hóa câu hỏi "Kết quả là gì" (Ví dụ: Ai chính xác sẽ là người rời bỏ dịch vụ vào tháng sau?), vươn tới đỉnh cao của sức mạnh dự báo bằng cách hy sinh một phần khả năng diễn giải để đổi lấy độ chính xác tuyệt đối. Cuốn giáo trình này không bắt người học phải chọn một trong hai con đường cực đoan đó. Thay vào đó, nó dung hòa và tích hợp cả hai lăng kính, tạo ra một tư duy phân tích toàn diện, nơi mà nhà khoa học dữ liệu vừa hiểu thấu đáo cơ chế cốt lõi sinh ra dữ liệu, lại vừa sở hữu năng lực dự báo mạnh mẽ vượt trội.

Tầm nhìn chiến lược này không phải là một ý tưởng ngẫu nhiên, mà nó được bắt nguồn trực tiếp từ triết lý và định hướng đào tạo tinh hoa của chương trình Economic Data Analytics (EDA) thuộc Khoa Toán Kinh tế (MFE), Đại học Kinh tế Quốc dân. Chúng tôi vô cùng tự hào khẳng định vị thế của một đại học tiên phong tại Việt Nam trong việc liên tục đổi mới, cập nhật và trang bị cho thế hệ trẻ những công cụ định lượng hiện đại nhất, tiệm cận một cách xuất sắc với các tiêu chuẩn học thuật và thực tiễn khắt khe trên toàn cầu. Đại học Kinh tế Quốc dân không chỉ là nơi truyền đạt những lý thuyết hàn lâm khô khan nằm ngoan ngoãn trên giấy, mà thực sự là một không gian học thuật năng động, nơi kiến thức được chuyển hóa mạnh mẽ thành những loại vũ khí sắc bén nhất để giải quyết các bài toán kinh tế phức tạp, đầy biến động của kỷ nguyên số. Thông qua chương trình EDA, Khoa MFE cam kết tạo ra một thế hệ chuyên gia phân tích dữ liệu ưu tú, những người có khả năng dẫn dắt sự thay đổi của nền kinh tế thay vì chỉ mệt mỏi chạy theo công nghệ mới. Việc biên soạn cuốn sách này chính là sự vật chất hóa rõ nét nhất của khát vọng to lớn đó, nhằm mục đích trực tiếp giúp sinh viên và người học thực hiện một bước nhảy vọt quan trọng trong nhận thức: chuyển đổi từ trạng thái học lý thuyết suông, rập khuôn máy móc các công thức, sang một tư duy mô hình hóa (modeling mindset) sắc sảo. Nhờ đó, họ sẽ có đủ bản lĩnh để bóc tách, cấu trúc hóa và đối mặt với bất kỳ một bài toán thực tiễn gai góc nào phát sinh từ đại dương dữ liệu bao la ngoài kia.

Chính vì mang đậm tính hàn lâm sâu sắc kết hợp chặt chẽ với sự thực chiến khốc liệt, đối tượng độc giả mà cuốn sách này hướng tới một cách có chủ đích là những sinh viên năm cuối xuất sắc, các học viên cao học, những nghiên cứu sinh đang khao khát tìm kiếm một phương pháp luận nghiên cứu định lượng vững chắc, và đặc biệt là các chuyên viên phân tích dữ liệu, chuyên viên quản trị rủi ro (Quants) đang phải đối mặt với áp lực làm việc khổng lồ tại các định chế tài chính, tập đoàn công nghệ lớn. Để có thể lĩnh hội trọn vẹn và đi hết hành trình tri thức đầy thử thách của cuốn sách này mà không bị gục ngã giữa chừng, người đọc bắt buộc phải trang bị cho mình một nền tảng tư duy định lượng vững chắc và các kỹ năng lập trình căn bản. Cụ thể, nền tảng Xác suất thống kê là một điều kiện tiên quyết mang tính sống còn và không thể thương lượng. Người học phải thực sự thấu hiểu sâu sắc bản chất triết học đằng sau giá trị p-value, nắm rõ ý nghĩa vật lý của phương sai, độ lệch chuẩn, và thấu cảm được cơ chế hoạt động của các phân phối xác suất ngẫu nhiên. Bởi vì, nếu thiếu đi lăng kính thống kê khắt khe này, mọi kết quả chạy ra từ máy tính sẽ chỉ là những con số vô hồn, đánh lừa thị giác và rất dễ dàng dẫn đến những diễn giải sai lầm, từ đó tạo ra những quyết định đầu tư mang tính thảm họa. Đồng thời, kỹ năng lập trình Python, đặc biệt là việc làm chủ sức mạnh của thư viện pandas để thao tác, biến đổi, nhào nặn và làm sạch dữ liệu, cũng là một yêu cầu tiên quyết. Trong thế giới thực tại các doanh nghiệp, dữ liệu không bao giờ nằm sẵn trong một định dạng hoàn hảo, sạch sẽ và ngăn nắp; chúng luôn chứa đựng đầy rẫy sự nhiễu loạn, các giá trị khuyết thiếu phức tạp, các yếu tố ngoại lai dị biệt và sự phi logic. Nếu không có một kỹ năng lập trình thao tác dữ liệu vững vàng, người phân tích sẽ mãi mãi mắc kẹt ở khâu tiền xử lý, chìm đắm trong các tập lệnh báo lỗi, và sẽ không bao giờ chạm tới được đỉnh cao nghệ thuật của việc xây dựng và tinh chỉnh các mô hình dự báo. Do đó, có thể nói một cách hình tượng rằng, toán học thống kê cung cấp linh hồn, còn Python cung cấp thể xác vật lý để chúng ta thực thi các ý tưởng phân tích phức tạp nhất một cách trọn vẹn.

Xuyên suốt toàn bộ các chương mục của cuốn sách này, triết lý sư phạm cốt lõi và bất di bất dịch mà chúng tôi kiên định theo đuổi là phương pháp tiếp cận dựa trên một Case Study thực tế duy nhất, liền mạch và mang tính End-to-End toàn diện. Thay vì đi theo lối mòn truyền thống là giới thiệu các thuật toán toán học bằng vô số những ví dụ nhỏ lẻ, vụn vặt và rời rạc, làm phân tán sự tập trung cũng như phá vỡ tư duy hệ thống của người học, chúng tôi tin tưởng tuyệt đối rằng giá trị thực sự của việc đào tạo thực chiến nằm ở việc cho phép người học tự mình trải nghiệm một luồng công việc (pipeline) trọn vẹn y hệt như một chuyên gia khoa học dữ liệu thực thụ đang làm việc tại một tập đoàn lớn. Bằng cách sử dụng một bộ dữ liệu ngân hàng duy nhất xuyên suốt tác phẩm, người học sẽ được trực tiếp chứng kiến toàn bộ vòng đời tiến hóa của dữ liệu: bắt đầu từ những dòng dữ liệu thô ráp, hỗn độn đầy rẫy lỗi kỹ thuật ở giai đoạn thu thập ban đầu, trải qua quá trình làm sạch đầy gian nan, xử lý các giá trị khuyết thiếu một cách tinh tế dựa trên xác suất, tiến đến việc trích xuất các đặc trưng tiềm ẩn bằng phương pháp phân tích thành phần chính PCA, sau đó phân cụm khách hàng chiến lược bằng thuật toán K-Means, và cuối cùng, thăng hoa với việc huấn luyện một siêu mô hình học máy Gradient Boosting như XGBoost để dự báo hành vi vỡ nợ hoặc rời bỏ của khách hàng. Sự liền mạch tuyệt đối này không chỉ giúp người học nắm bắt được cơ chế hoạt động toán học độc lập của từng thuật toán riêng lẻ, mà quan trọng hơn gấp ngàn lần, nó khắc sâu vào tâm trí họ sự liên kết chặt chẽ mang tính hữu cơ giữa tất cả các khâu trong đường ống dữ liệu. Ở đó, người học sẽ nhận ra một bài học cay đắng nhưng vô giá: sự cẩu thả và hời hợt ở bước tiền xử lý ban đầu sẽ lập tức phải trả giá đắt bằng sự sụp đổ hoàn toàn của toàn bộ hệ thống dự báo ở giai đoạn cuối cùng. Cuốn sách này không hề hứa hẹn mang đến một con đường học tập dễ dàng, nhàn hạ hay đầy màu hồng, nhưng nó cam kết bằng mọi giá sẽ mang lại cho bạn một tư duy phân tích toàn diện, sâu sắc, sắc bén và sẵn sàng đương đầu với mọi thách thức phức tạp nhất của thế giới dữ liệu đa chiều.
