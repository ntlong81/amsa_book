# TÀI LIỆU THAM KHẢO (BIBLIOGRAPHY)

Trang này tổng hợp toàn bộ danh mục tài liệu chuẩn (textbooks) và các công trình khoa học kinh điển được trích dẫn xuyên suốt giáo trình, được phân chia theo từng chương học để tiện tra cứu và đối chiếu học thuật.

## 1. Tài liệu đối chiếu chính (Core Textbooks)

* **Hair, J. F., Black, W. C., Babin, B. J., & Anderson, R. E. (2018).** *Multivariate Data Analysis* (8th Edition). Cengage Learning.
  * *Vai trò:* Giáo trình lý thuyết chính cho các phương pháp phân tích đa biến, từ chuẩn bị dữ liệu đến cấu trúc SEM.
* **Johnson, R. A., & Wichern, D. W. (2007).** *Applied Multivariate Statistical Analysis* (6th Edition). Pearson Prentice Hall.
  * *Vai trò:* Tài liệu đối chiếu toán học và đại số tuyến tính đa chiều nền tảng.
* **Rencher, A. C., & Christensen, W. F. (2012).** *Methods of Multivariate Analysis* (3rd Edition). John Wiley & Sons.
  * *Vai trò:* Tài liệu phân tích sâu về các phép chứng minh ma trận, phân phối chuẩn đa biến và phân tích biệt thức.
* **James, G., Witten, D., Hastie, T., & Tibshirani, R. (2021).** *An Introduction to Statistical Learning: with Applications in Python*. Springer.
  * *Vai trò:* Cầu nối thực hành mô hình hóa dự báo, Regularization và Machine Learning bằng Python.
* **Hastie, T., Tibshirani, R., & Friedman, J. (2009).** *The Elements of Statistical Learning: Data Mining, Inference, and Prediction* (2nd Edition). Springer Science & Business Media.
  * *Vai trò:* Tài liệu nâng cao về triết lý đánh đổi Độ chệch - Phương sai (Bias-Variance Tradeoff) và các thuật toán học máy nâng cao.
* **McKinney, W. (2022).** *Python for Data Analysis* (3rd Edition). O'Reilly Media.
  * *Vai trò:* Hướng dẫn chuẩn mực cho các thao tác xử lý dữ liệu với pandas và numpy.

## 2. Tài liệu học thuật theo từng chương

### Chương 1: Cơ sở Toán học và Tổng quan về Phân tích Đa biến

* **Stevens, S. S. (1946).** *On the theory of scales of measurement*. Science, 103(2684), 677-680.
  * *Nội dung:* Công trình nền tảng định hình 4 thang đo dữ liệu (Nominal, Ordinal, Interval, Ratio).
* **Mahalanobis, P. C. (1936).** *On the generalized distance in statistics*. Proceedings of the National Institute of Sciences of India, 2(1), 49-55.
  * *Nội dung:* Đề xuất khoảng cách thống kê Mahalanobis để khử tương quan và đồng bộ hóa đơn vị đo.
* **Mardia, K. V. (1970).** *Measures of multivariate skewness and kurtosis with applications*. Biometrika, 57(3), 519-530.
  * *Nội dung:* Phát triển kiểm định độ lệch và độ nhọn đa biến cho giả định chuẩn.
* **Henze, N., & Zirkler, B. (1990).** *A class of invariant consistent tests for multivariate normality*. Communications in Statistics-Theory and Methods, 19(10), 3595-3617.
  * *Nội dung:* Xây dựng thống kê kiểm định tính chuẩn đa biến tối ưu dựa trên hàm đặc trưng.

### Chương 2: Chuẩn bị và Làm sạch Dữ liệu

* **Rubin, D. B. (1976).** *Inference and missing data*. Biometrika, 63(3), 581-592.
  * *Nội dung:* Định nghĩa và phân loại 3 cơ chế khuyết dữ liệu (MCAR, MAR, MNAR).
* **Little, R. J., & Rubin, D. B. (2019).** *Statistical Analysis with Missing Data* (3rd Edition). John Wiley & Sons.
  * *Nội dung:* Khung lý thuyết và các phương pháp xử lý dữ liệu thiếu thực tế.

### Chương 3: Tư duy Mô hình hóa Đa biến

* **Pearl, J. (2009).** *Causality: Models, Reasoning, and Inference* (2nd Edition). Cambridge University Press.
  * *Nội dung:* Lý thuyết đồ thị nhân quả DAGs, định lý Backdoor path và cơ chế suy diễn nhân quả.
* **Pearl, J., Glymour, M., & Jewell, N. P. (2016).** *Causality in Statistics: A Primer*. John Wiley & Sons.
  * *Nội dung:* Tài liệu giới thiệu cơ bản về sự phân biệt giữa suy diễn nhân quả và dự báo dữ liệu thuần túy.

### Chương 4: Phân tích thành phần chính (PCA) & Nhân tố Khám phá (EFA)

* **Pearson, K. (1901).** *On lines and planes of closest fit to systems of points in space*. The London, Edinburgh, and Dublin Philosophical Magazine and Journal of Science, 2(11), 559-572.
  * *Nội dung:* Công trình đầu tiên mô tả nền tảng hình học của PCA thông qua cực đại hóa phương sai.
* **Spearman, C. (1904).** *"General Intelligence," objectively determined and measured*. The American Journal of Psychology, 15(2), 201-292.
  * *Nội dung:* Đặt nền móng cho EFA bằng việc đo lường các khái niệm ẩn như trí thông minh.
* **Kaiser, H. F. (1958).** *The varimax criterion for analytic rotation in factor analysis*. Psychometrika, 23(3), 187-200.
  * *Nội dung:* Phép xoay Varimax giúp tối ưu hóa khả năng giải thích cấu trúc nhân tố.
* **Kaiser, H. F., & Rice, J. (1974).** *Little jiffy, mark IV*. Educational and Psychological Measurement, 34(1), 111-117.
  * *Nội dung:* Định nghĩa chỉ số KMO (Kaiser-Meyer-Olkin) đo lường tính tương thích của dữ liệu với phân tích nhân tố.

### Chương 5: Phân tích Cụm và Khám phá Cấu trúc Nhóm

* **MacQueen, J. (1967).** *Some methods for classification and analysis of multivariate observations*. In Proceedings of the Fifth Berkeley Symposium on Mathematical Statistics and Probability (Vol. 1, No. 14, pp. 281-297).
  * *Nội dung:* Thuật toán phân cụm phi phân cấp K-Means kinh điển.
* **Dempster, A. P., Laird, N. M., & Rubin, D. B. (1977).** *Maximum likelihood from incomplete data via the EM algorithm*. Journal of the Royal Statistical Society: Series B (Methodological), 39(1), 1-22.
  * *Nội dung:* Thuật toán EM làm nền tảng cho việc tối ưu phân cụm xác suất bằng Gaussian Mixture Models (GMM).
* **Rousseeuw, P. J. (1987).** *Silhouettes: a graphical aid to the interpretation and validation of cluster analysis*. Journal of Computational and Applied Mathematics, 20, 53-65.
  * *Nội dung:* Đề xuất chỉ số Silhouette đánh giá mật độ và khoảng cách giữa các cụm.

### Chương 6: Phân tích Tương quan, ANOVA và Hồi quy Tuyến tính Bội

* **Hoerl, A. E., & Kennard, R. W. (1970).** *Ridge regression: Biased estimation for nonorthogonal problems*. Technometrics, 12(1), 55-67.
  * *Nội dung:* Giải pháp Ridge Regression sử dụng hàm phạt L2 để kiểm soát đa cộng tuyến.
* **Tibshirani, R. (1996).** *Regression shrinkage and selection via the lasso*. Journal of the Royal Statistical Society: Series B (Methodological), 58(1), 267-288.
  * *Nội dung:* Kỹ thuật Lasso Regression sử dụng hàm phạt L1 phục vụ co rút và lựa chọn đặc trưng đồng thời.

### Chương 7: Phân tích Biệt số và Hồi quy Logistic

* **Fisher, R. A. (1936).** *The use of multiple measurements in taxonomic problems*. Annals of Eugenics, 7(2), 179-188.
  * *Nội dung:* Nền tảng toán học của thuật toán phân tích biệt thức tuyến tính (LDA).
* **Cox, D. R. (1958).** *The regression analysis of binary sequences*. Journal of the Royal Statistical Society: Series B (Methodological), 20(2), 215-232.
  * *Nội dung:* Khái quát hóa mô hình hồi quy logistic ứng dụng cho các biến mục tiêu nhị phân.
* **Chawla, N. V., Bowyer, K. W., Hall, L. O., & Kegelmeyer, W. P. (2002).** *SMOTE: synthetic minority over-sampling technique*. Journal of Artificial Intelligence Research, 16, 321-357.
  * *Nội dung:* Thuật toán tạo mẫu tổng hợp SMOTE để xử lý mất cân bằng dữ liệu.

### Chương 8: Phân tích nhân tố khẳng định (CFA) & Mô hình Cấu trúc (SEM)

* **Jöreskog, K. G. (1969).** *A general approach to confirmatory maximum likelihood factor analysis*. Psychometrika, 34(2), 183-202.
  * *Nội dung:* Phương pháp maximum likelihood cho mô hình CFA.
* **Fornell, C., & Larcker, D. F. (1981).** *Evaluating structural equation models with unobservable variables and measurement error*. Journal of Marketing Research, 18(1), 39-50.
  * *Nội dung:* Tiêu chuẩn đánh giá độ tin cậy AVE, CR và tính hội tụ/phân biệt của thang đo.
* **Bollen, K. A. (1989).** *Structural Equations with Latent Variables*. John Wiley & Sons.
  * *Nội dung:* Sách giáo khoa kinh điển đặt nền móng lý thuyết vững chắc cho SEM.

### Chương 9: Chuỗi thời gian Đa biến (VAR/VECM)

* **Sims, C. A. (1980).** *Macroeconomics and reality*. Econometrica, 48(1), 1-48.
  * *Nội dung:* Đề xuất mô hình tự hồi quy vector (VAR) khắc phục tính nội sinh trong kinh tế vĩ mô.
* **Engle, R. F., & Granger, C. W. (1987).** *Co-integration and error correction: Representation, estimation, and testing*. Econometrica, 55(2), 251-276.
  * *Nội dung:* Khái niệm đồng liên kết và lời giải VECM kết hợp tác động ngắn hạn và cân bằng dài hạn.
* **Johansen, S. (1988).** *Statistical analysis of cointegration vectors*. Journal of Economic Dynamics and Control, 12(2-3), 231-254.
  * *Nội dung:* Kiểm định Johansen xác định số lượng vector đồng liên kết đa chiều.
* **Hamilton, J. D. (1994).** *Time Series Analysis*. Princeton University Press.
  * *Nội dung:* Giáo trình chuẩn mực và toàn diện nhất về toán học chuỗi thời gian.

### Chương 10: Tích hợp Machine Learning và XAI vào Phân tích Đa biến

* **Breiman, L. (2001).** *Random forests*. Machine Learning, 45(1), 5-32.
  * *Nội dung:* Thuật toán rừng ngẫu nhiên áp dụng kỹ thuật gộp mẫu ngẫu nhiên (Bagging).
* **Chen, T., & Guestrin, C. (2016).** *XGBoost: A scalable tree boosting system*. In Proceedings of the 22nd ACM SIGKDD International Conference on Knowledge Discovery and Data Mining (pp. 785-794).
  * *Nội dung:* Kiến trúc và toán học thuật toán Gradient Boosting nâng cao (XGBoost).
* **Ribeiro, M. T., Singh, S., & Guestrin, C. (2016).** *"Why should I trust you?": Explaining the predictions of any classifier*. In Proceedings of the 22nd ACM SIGKDD International Conference on Knowledge Discovery and Data Mining (pp. 1135-1144).
  * *Nội dung:* Giới thiệu mô hình LIME giải thích dự báo cục bộ của mô hình hộp đen.
* **Lundberg, S. M., & Lee, S.-I. (2017).** *A unified approach to interpreting model predictions*. Advances in Neural Information Processing Systems, 30.
  * *Nội dung:* Đề xuất giá trị SHAP dựa trên lý thuyết trò chơi của Shapley.
* **Molnar, C. (2022).** *Interpretable Machine Learning: A Guide for Making Black Box Models Explainable* (2nd Edition).
  * *Nội dung:* Sách giáo khoa lý thuyết và thực hành chuẩn mực về khả năng giải thích của học máy.
